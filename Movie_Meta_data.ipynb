```python
## Checcking if variables are normally distributed or not
def plot_histplot(df):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(df.columns):
        plt.subplot(4,4,i+1)
        plt.hist(df[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
### Dealing missing values
def fill_missing_values(df,strategy='mean'):
    num_cols = df.select_dtypes(include=['int64','float64']).columns
    cat_cols = df.select_dtypes(include=['object','category']).columns
    if strategy == 'mean':
        df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
    elif strategy == 'median':
        df[num_cols] = df[num_cols].fillna(df[num_cols].median())
    elif strategy == 'mode':
        df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
    return df
```

```python
### Dealing with outliers
def plot_boxplot(df):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(df.columns):
        plt.subplot(4,4,i+1)
        plt.boxplot(df[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
plot_boxplot(cont_df2)
```

```python
## Removing the outliers
def remove_outliers(df):
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)

        IQR = Q3 - Q1
        ### Setting boundaries
        lower_boundary = Q1 - IQR * 1.5
        upper_boundary = Q3 + IQR * 1.5

        df[col] = np.where(df[col] < lower_boundary,lower_boundary,df[col])
        df[col] = np.where(df[col] > upper_boundary,upper_boundary,df[col])
```

```python
remove_outliers(cont_df2)
```
