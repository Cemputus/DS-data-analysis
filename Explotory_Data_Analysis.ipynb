```python
## Checking for missing values
df.isna().sum()
```

```python
cont_df = df.select_dtypes(exclude='object')
cont_df
```

```python
### PLot if it had some missing values
def plot_histplot(df):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(df.columns):
        plt.subplot(3,3,i+1)
        plt.hist(df[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
plot_histplot(cont_df)
```

```python
def plot_scattertplot(df):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(df.columns):
        plt.subplot(3,3,i+1)
        sns.scatterplot(df[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
plot_scattertplot(cont_df)
```

```python
### cheching for outliers
def plot_boxtplot(df):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(df.columns):
        plt.subplot(3,3,i+1)
        plt.boxplot(df[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
plot_boxtplot(cont_df)
```
